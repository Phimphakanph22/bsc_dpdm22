```python
assessor2['ประเมินโดย1'] = assessor2['post id'].map(assessor1.set_index('post id')['ประเมินราคาซื้อ-ขาย']);
assessor2
```

```python
assessor2[['post id','ราคาทรัพย์']]
```

```python
assessor2.rename(columns={'ราคาทรัพย์':'ราคาตั้งขาย'},inplace=True);
assessor2
```

```python
assessor2.rename(columns={'ประเมินราคาซื้อ-ขาย':'ประเมินโดย2'},inplace=True);
assessor2
```

```python
assessor = assessor2[['post id','ราคาตั้งขาย','ประเมินโดย1','ประเมินโดย2']]
assessor
```

```python
assessor.isnull().any()
```

```python
assessorall = assessor.dropna(subset=['ประเมินโดย2'])
assessorall
```

```python
#from sklearn.metrics import mean_absolute_error
#Y_actual = [1,2,3,4,5]
#Y_Predicted = [1,2.5,3,4.1,4.9]
#mape = mean_absolute_error(Y_actual, Y_Predicted)*100
#print(mape)
```

```python
Y_actual = assessorall[['ราคาตั้งขาย']]
Y_Predicted1 = assessorall[['ประเมินโดย1']]
Y_Predicted2 = assessorall[['ประเมินโดย2']]
```
