```python
from sklearn import datasets

iris_data = datasets.load_iris()
```

```python
X = iris_data['data']

y = iris_data['target']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, KFold
```

```python
model = DecisionTreeClassifier()

# Define the k-fold cross validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Train and evaluate the model using 10-fold cross validation
scores = cross_val_score(model, X, y, cv=kf)

# Print the average accuracy and standard deviation across all folds
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
model = DecisionTreeClassifier(max_depth=2)

# Define the k-fold cross validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Train and evaluate the model using 10-fold cross validation
scores = cross_val_score(model, X, y, cv=kf)

# Print the average accuracy and standard deviation across all folds
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

# HW6 วนลูปทำ 10-fold cross validation

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import LogisticRegression
```

```python
# แบ่งตาราง 2 ตาราง Fold = 1, 2-10
iris_table1 = shuffle_iris[shuffle_iris['fold']==1]
iris_table2 = shuffle_iris[shuffle_iris['fold'] > 1]

# For loop Validation
x = iris_table2.drop(['Lable','fold'], axis=1) #Pure data to validation
y = iris_table2['Lable']
kf = KFold(n_splits=9, shuffle=True, random_state=6)
```

```python
# Create a Decision Tree classifier object
dt = DecisionTreeClassifier(random_state=6)

# Create a 10-fold cross-validation object
kf = KFold(n_splits=10, shuffle=True, random_state=6)

# Perform 10-fold cross-validation
scores = cross_val_score(dt, x, y, cv=kf)
```

```python
# Loop through the 10 folds
for fold, (train_index, test_index) in enumerate(kf.split(x)):
    print(f"Fold {fold + 1}")
    
    # Get the training and test sets for this fold
    X_train, y_train = x.iloc[train_index], y.iloc[train_index]
    X_test, y_test = x.iloc[test_index], y.iloc[test_index]
    
    # Train the model on the training set
    dt.fit(X_train, y_train)
    
    # Evaluate the model on the test set
    score = dt.score(X_test, y_test)
    print(f"Accuracy: {score}")
    
    # Get the cross-validation score for this fold
    cv_score = cross_val_score(dt, X_train, y_train, cv=kf).mean()
    print(f"Cross-validation score: {cv_score}")
```
